```python
import numpy as np 
import pandas as pd 
import os
import gc
import psutil
```

```python
print(os.listdir('E:\\kaggle\\Avito Demand Prediction Challenge'))
```

```python
#columns=['item_id','user_id','activation_date']
#columns=['item_id','user_id']
path='E:\\kaggle\\Avito Demand Prediction Challenge\\'
test=pd.read_csv(path+'test.csv',header=0)
test['deal_probability']=-1
train=pd.read_csv(path+'train.csv',header=0)
train['price'] = train['price'].fillna(0).astype('float32')
test.head()
```

```python
category=train.groupby(['parent_category_name','category_name']).deal_probability.mean().reset_index()
category.columns.values[2]='deal_prob_avg'
category2=train.groupby(['parent_category_name']).deal_probability.mean().reset_index()
category2.columns.values[1]='deal_prob_avg_par'
category=pd.merge(category, category2, on=['parent_category_name'],how='left')
del category2
gc.collect()
category
```

```python
data=pd.concat([train,test],ignore_index=True)

cate=data.groupby(['parent_category_name','category_name']).item_id.nunique().reset_index()
cate.columns.values[2]='count_avg'
cate2=train.groupby(['parent_category_name']).item_id.nunique().reset_index()
cate2.columns.values[1]='count_avg_par'
cate=pd.merge(cate, cate2, on=['parent_category_name'],how='left')
del cate2
gc.collect()
cate
```

```python
category=pd.merge(category,cate,on=['parent_category_name','category_name'],how='left')
category.to_csv('classfeature.csv', index=False)
category
```
