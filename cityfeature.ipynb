```python
columns=['item_id','region','city']
path='E:\\kaggle\\Avito Demand Prediction Challenge\\'
train=pd.read_csv(path+'train.csv',header=0,skipinitialspace=True, usecols=columns)
test=pd.read_csv(path+'test.csv',header=0,skipinitialspace=True, usecols=columns)
data=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()

data=data.groupby(['region','city'])['item_id'].size().reset_index()
data.columns.values[2] = 'count'

#population=pd.read_csv(path+'city_population_wiki_v3.csv',header=0)
#data=pd.merge(data, population, on=['city'],how='left')

location=pd.read_csv(path+'avito_region_city_features.csv',header=0)
location=location.drop(['city_region','region_id','city_region_id'], axis=1)
location.columns.values[4] = 'cluster_5'
location.columns.values[5] = 'cluster_10'
location.columns.values[6] = 'cluster_20'

data=pd.merge(data, location, on=['region','city'],how='left')

data
```

```python
from math import sin, cos, sqrt, atan2, radians
def distance(lat1,lon1,lat2,lon2):
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    
    return distance
```

```python
regionlist=list(np.unique(data['region']))
cityname=[]
regionname=[]
ship_count=[]
cus_count=[]
for region in regionlist:
    temp=data[data['region']==region].reset_index(drop=True)
    for i in range(len(temp)):
        cityname.append(temp.loc[i,'city'])
        regionname.append(region)
        cus=0
        ship=0
        count=0
        for j in range(len(temp)):
            r=distance(temp.loc[i,'latitude'],temp.loc[i,'longitude'],temp.loc[j,'latitude'],temp.loc[j,'longitude'])  
            if r<=25:
                index=1
            else:
                index=(r/25.0)**1.1
            cus=cus+temp.loc[j,'count']/index
            
            count=count+temp.loc[j,'count']
            if r<=10:
                r=10
            ship=ship+temp.loc[j,'count']*r
        
        ship=ship/count
        ship_count.append(ship)
        cus_count.append(cus)
    print (region)

result = {'region':regionname,'city':cityname,'ship':ship_count,'cus':cus_count}
result = pd.DataFrame(data=result)
data=pd.merge(data, result, on=['region','city'],how='left')
del result
gc.collect()
data
```

```python
region=pd.read_csv(path+'region.csv',header=0)
regionname=[]
ship_count=[]
cus_count=[]
for i in range(len(region)):
    regionname.append(region.loc[i,'region'])
    cus=0
    ship=0
    count=0
    for j in range(len(region)):
        r=distance(region.loc[i,'r_lat'],region.loc[i,'r_lon'],region.loc[j,'r_lat'],region.loc[j,'r_lon'])  
        if r<=25:
            index=1
        else:
            index=(r/25.0)**1.1
        cus=cus+region.loc[j,'r_count']/index
            
        count=count+region.loc[j,'r_count']
        if r<=10:
            r=10
        ship=ship+region.loc[j,'r_count']*r
        
    ship=ship/count
    ship_count.append(ship)
    cus_count.append(cus)
result = {'region':regionname,'r_ship':ship_count,'r_cus':cus_count}
result = pd.DataFrame(data=result)
data=pd.merge(data, result, on=['region'],how='left')
del result
gc.collect()
data
```
