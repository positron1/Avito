```python
columns=['item_id','user_id']
path='E:\\kaggle\\Avito Demand Prediction Challenge\\'
test=pd.read_csv(path+'test_active.csv',header=0,skipinitialspace=True, usecols=columns)
train=pd.read_csv(path+'train_active.csv',header=0,skipinitialspace=True, usecols=columns)
data=pd.concat([train,test],ignore_index=True)
data['temp']=0
data=data.groupby(['item_id','user_id']).temp.size().reset_index()
data.columns.values[2]='temp'
data=data.drop(['temp'], axis=1)
del train,test
gc.collect()

columns=['item_id','activation_date']
test=pd.read_csv(path+'periods_test.csv',header=0,skipinitialspace=True, usecols=columns)
train=pd.read_csv(path+'periods_train.csv',header=0,skipinitialspace=True, usecols=columns)
period=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()
period
```

```python
period=period.groupby(['item_id']).activation_date.size().reset_index()
period.columns.values[1]='edit_time'
gc.collect()
data=pd.merge(data, period, on=['item_id'],how='left')
del period
gc.collect()
data=data.fillna(1)
gc.collect()
data['edit_time']=data['edit_time']-1

sumcount=data.groupby(['user_id']).edit_time.sum().reset_index()
sumcount.columns.values[1]='edit_time'

count=data.groupby(['user_id']).item_id.nunique().reset_index()
count.columns.values[1]='count'

data=pd.merge(sumcount, count, on=['user_id'],how='left')
data['edit_time']=data['edit_time']/data['count']
data=data.drop(['count'], axis=1)
gc.collect()
data['edit_time']=round(data['edit_time'],3)

data.to_csv('usereditnum.csv', index=False)
data
```

```python
columns=['item_id','user_id']
path='E:\\kaggle\\Avito Demand Prediction Challenge\\'
test=pd.read_csv(path+'test_active.csv',header=0,skipinitialspace=True, usecols=columns)
train=pd.read_csv(path+'train_active.csv',header=0,skipinitialspace=True, usecols=columns)
data=pd.concat([train,test],ignore_index=True)
data['temp']=0
data=data.groupby(['item_id','user_id']).temp.size().reset_index()
data.columns.values[2]='temp'
data=data.drop(['temp'], axis=1)
del train,test
gc.collect()

columns=['item_id','date_from','date_to']
test=pd.read_csv(path+'periods_test.csv',header=0,skipinitialspace=True, usecols=columns)
train=pd.read_csv(path+'periods_train.csv',header=0,skipinitialspace=True, usecols=columns)
period=pd.concat([train,test],ignore_index=True)
del train,test
gc.collect()
period
```

```python
timeorigin=pd.to_datetime('2017-03-01')
period['date_from']=pd.to_datetime(period['date_from'], format="%Y-%m-%d")
period['date_to']=pd.to_datetime(period['date_to'], format="%Y-%m-%d")
period['dis_days']=(period['date_to']-period['date_from']).dt.days.astype('uint')
period=period.drop(['date_from','date_to'], axis=1)

period=period.groupby(['item_id']).dis_days.sum().reset_index()
period.columns.values[1]='dis_days'
gc.collect()
data=pd.merge(data, period, on=['item_id'],how='left')
del period
gc.collect()
data
```

```python
data=data.fillna(1)
gc.collect()

sumcount=data.groupby(['user_id']).dis_days.sum().reset_index()
sumcount.columns.values[1]='dis_days'

count=data.groupby(['user_id']).item_id.nunique().reset_index()
count.columns.values[1]='count'

data=pd.merge(sumcount, count, on=['user_id'],how='left')
data['dis_days']=data['dis_days']/data['count']
data=data.drop(['count'], axis=1)
gc.collect()
data['dis_days']=round(data['dis_days'],3)

data.to_csv('userdisplaydays.csv', index=False)
data
```
